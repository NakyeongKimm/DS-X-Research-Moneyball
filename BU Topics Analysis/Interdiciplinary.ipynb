# 2021

```python
import pandas as pd
import numpy as np

column_indices_to_force_str = [21, 34, 53, 54, 66, 67, 108, 112, 113, 175, 176, 177, 178, 179, 180, 181, 182]
# If your CSV has headers, you'd use the column names instead of indices.

dtype_specification = {col: str for col in column_indices_to_force_str}

df_2021 = pd.read_csv('/content/drive/MyDrive/hack/2021_t15.csv', dtype=dtype_specification)
```

```python
cumulative_distinct_list = []

for i in range(len(df_2021)):
    current_distinct_values = set()

    cell_data = df_2021['topics.domain.display_name'].iloc[i]

    a = [domain.strip() for domain in cell_data.split('|')]

    current_distinct_values.update(a)
    cumulative_distinct_list.append(list(current_distinct_values))
```

```python
df_2021['cumulative_distinct_domains'] = cumulative_distinct_list
```

```python
from itertools import combinations
from collections import Counter

domain_column = df_2021['cumulative_distinct_domains']

all_pairs = []
for domain_list in domain_column:
  if isinstance(domain_list, set):
    domain_list = list(domain_list)
  if len(domain_list) < 2:
    continue
  for pair in combinations(sorted(domain_list), 2):
    all_pairs.append(pair)

co_occurence_counts = Counter(all_pairs)
```

```python
df_co_occurrence_2021 = pd.DataFrame(co_occurence_counts.items(), columns=['Domain Pair', 'Count'])
df_co_occurrence_2021 = df_co_occurrence_2021.sort_values(by='Count', ascending=False).reset_index(drop=True)
```

```python
print(df_co_occurrence_2021.head(15))
```

```python
total_count_sum = df_co_occurrence_2021['Count'].sum()
df_co_occurrence_2021['Ratio'] = (df_co_occurrence_2021['Count'] / total_count_sum) * 100
final_df = df_co_occurrence_2021.sort_values(by='Ratio', ascending=False).reset_index(drop=True)
print(final_df.head())
```

```python
final_df['Pair Label'] = final_df['Domain Pair'].apply(lambda x: f"{x[0]} & {x[1]}")

plt.figure(figsize=(12, 7))

sns.barplot(
    x='Ratio',
    y='Pair Label',
    data=final_df,
    palette='viridis'
)

plt.title('Percentage Distribution of Co-occurrence Pairs (Year 2021)', fontsize=16)
plt.xlabel('Ratio (%)', fontsize=12)
plt.ylabel('Domain Pair', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.6) # x축 그리드 추가
plt.show()
```

```python
df_list = [df_co_occurrence_2021, df_co_occurrence_2022, df_co_occurrence_2023,
           df_co_occurrence_2024, df_co_occurrence_2025]

combined_df = pd.concat(df_list, ignore_index = True)
total_co_occurence = combined_df.groupby('Domain Pair')['Count'].sum()
final_df = total_co_occurence.reset_index()
final_df = final_df.sort_values(by='Count', ascending=False)
print(final_df.head())
```

```python
total_count_sum = final_df['Count'].sum()
final_df['Ratio'] = (final_df['Count'] / total_count_sum) * 100
final_df = final_df.sort_values(by='Ratio', ascending=False).reset_index(drop=True)
print(final_df.head())
```

```python
final_df['Pair Label'] = final_df['Domain Pair'].apply(lambda x: f"{x[0]} & {x[1]}")

plt.figure(figsize=(12, 7))

sns.barplot(
    x='Ratio',
    y='Pair Label',
    data=final_df,
    palette='viridis'
)

plt.title('Percentage Distribution of Co-occurrence Pairs (Total 5 Years)', fontsize=16)
plt.xlabel('Ratio (%)', fontsize=12)
plt.ylabel('Domain Pair', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

for index, row in final_df.iterrows():
    domain1, domain2 = row['Domain Pair']
    count = row['Count']

    G.add_edge(domain1, domain2, weight=count)

plt.figure(figsize=(12, 10))

pos = nx.spring_layout(G, k=0.4, iterations=50, seed=42)

edge_weights = [G[u][v]['weight'] for u, v in G.edges()]
max_weight = max(edge_weights)
edge_widths = [w / max_weight * 9 + 1 for w in edge_weights]

node_sizes = [G.degree(node) * 1400 for node in G.nodes()]

nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=node_sizes, alpha=0.9)

nx.draw_networkx_edges(
    G, pos,
    width=edge_widths,
    edge_color='gray',
    alpha=0.6
)

nx.draw_networkx_labels(G, pos, font_size=12, font_color='black', font_weight='bold')

edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=14, label_pos=0.5, font_color='red')

plt.title("Domain Co-occurrence Network (Interdisciplinary Connections)", size=16, pad=20)
plt.axis('off')
plt.show()
```

```python
import pandas as pd
import numpy as np

all_domains = set()
for pair in final_df['Domain Pair']:
    all_domains.update(pair)
domain_list = sorted(list(all_domains))

n = len(domain_list)
co_occurrence_matrix = np.zeros((n, n), dtype=int)
matrix_df = pd.DataFrame(co_occurrence_matrix, index=domain_list, columns=domain_list)

for index, row in final_df.iterrows():
    dom1, dom2 = row['Domain Pair']
    count = row['Count']

    matrix_df.loc[dom1, dom2] = count
    matrix_df.loc[dom2, dom1] = count
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 7))

sns.heatmap(
    matrix_df,
    annot=True,
    fmt="d",
    cmap="YlGnBu",
    linewidths=.5,
    cbar_kws={'label': 'Co-occurrence Count'}
)

plt.title('Interdisciplinary Co-occurrence Heatmap', fontsize=16)
plt.show()
```
